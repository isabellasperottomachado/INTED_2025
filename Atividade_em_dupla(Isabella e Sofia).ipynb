```python
#1 -
def remove_duplicatas_mantendo_ordem():
    '''
    Cria uma lista de 15 números inteiros digitados pelo usuário e remove as duplicatas.
    '''
    numeros = []

    for i in range(15):
        while True:
            try:
                num = int(input(f"Digite o {i+1}º número inteiro: "))
                numeros.append(num)
                break
            except ValueError:
                print("Entrada inválida. Por favor, digite um número inteiro.")

    lista_sem_duplicatas = []
    numeros_vistos = set()

    for num in numeros:
        if num not in numeros_vistos:
            lista_sem_duplicatas.append(num)
            numeros_vistos.add(num)

    print("Lista original:", numeros)
    print("Lista sem duplicatas (mantendo a ordem):", lista_sem_duplicatas)

remove_duplicatas_mantendo_ordem()
```

```python
# 5 -
import math

def eh_primo(numero):
    """
    Verifica se um número é primo.
    """
    if numero <= 1:
        return False
    if numero == 2:
        return True
    if numero % 2 == 0:
        return False
    for i in range(3, int(math.sqrt(numero)) + 1, 2):
        if numero % i == 0:
            return False
    return True

def main():
    """
    Função principal que gerencia o fluxo do programa.
    """
    lista_principal = []

    print("Digite 10 números inteiros:")
    for i in range(10):
        while True:
            try:
                numero = int(input(f"Número {i+1}: "))
                lista_principal.append(numero)
                break
            except ValueError:
                print("Entrada inválida. Por favor, digite um número inteiro.")

    lista_primos = []

    for numero in lista_principal:
        if eh_primo(numero):
            lista_primos.append(numero)

    print("\n--- Resultados ---")
    print("Lista original:", lista_principal)
    print("Lista de números primos:", lista_primos)

if __name__ == "__main__":
    main()
```

```python
# 6 -
def verificar_palindromo():
    """
    Lê 8 elementos do usuário e verifica se a lista é um palíndromo.
    """
    lista = []

    print("Digite 8 elementos para preencher a lista:")
    for i in range(8):
        elemento = input(f"Digite o {i+1}º elemento: ")
        lista.append(elemento)

    lista_invertida = lista[::-1]

    if lista == lista_invertida:
        print(f"\nLista original: {lista}")
        print("A lista é um palíndromo!")
    else:
        print(f"\nLista original: {lista}")
        print("A lista não é um palíndromo.")

verificar_palindromo()
```

```python
# 7 -
import random

def encontrar_intersecao():
    """
    Cria duas listas com 10 números aleatórios, encontra a interseção
    e exibe as três listas.
    """
    lista1 = [random.randint(1, 100) for _ in range(10)]

    lista2 = [random.randint(1, 100) for _ in range(10)]

    print("Primeira lista:", lista1)
    print("Segunda lista:", lista2)

    conjunto1 = set(lista1)
    conjunto2 = set(lista2)

    intersecao_conjunto = conjunto1.intersection(conjunto2)

    lista_intersecao = list(intersecao_conjunto)

    print("Lista de interseção:", lista_intersecao)

encontrar_intersecao()
```

```python
#9 -
def somatorio_acumulado():
    """
    Solicita 10 números ao usuário, cria uma lista com o somatório
    acumulado e exibe as duas listas.
    """
    lista_original = []

    print("Digite 10 números inteiros:")
    for i in range(10):
        while True:
            try:
                numero = int(input(f"Número {i+1}: "))
                lista_original.append(numero)
                break
            except ValueError:
                print("Entrada inválida. Por favor, digite um número inteiro.")

    lista_acumulada = []
    soma_atual = 0
    for elemento in lista_original:
        soma_atual += elemento
        lista_acumulada.append(soma_atual)

    print("\n--- Resultados ---")
    print("Lista original:", lista_original)
    print("Lista com somatório acumulado:", lista_acumulada)

somatorio_acumulado()
```

```python
#11 -
def criar_matriz_e_somar_diagonal():
    """
    Lê uma matriz 4x4 do usuário, exibe a matriz e a soma da diagonal secundária.
    """
    tamanho = 4
    matriz = []

    print("Digite os valores para a matriz 4x4:")
    for i in range(tamanho):
        linha = []
        for j in range(tamanho):
            while True:
                try:
                    valor = int(input(f"Digite o valor para a posição [{i+1}][{j+1}]: "))
                    linha.append(valor)
                    break
                except ValueError:
                    print("Entrada inválida. Por favor, digite um número inteiro.")
        matriz.append(linha)

    print("\n--- Matriz preenchida ---")
    for linha in matriz:
        print(linha)

    soma_diagonal_secundaria = 0
    for i in range(tamanho):
        soma_diagonal_secundaria += matriz[i][tamanho - 1 - i]

    print(f"\nA soma dos elementos da diagonal secundária é: {soma_diagonal_secundaria}")

criar_matriz_e_somar_diagonal()
```

```python
#12 -

import random

def calcular_media_linhas_matriz():
    """
    Gera uma matriz 5x5 com números aleatórios, calcula a média de cada
    linha e armazena os resultados em uma nova lista.
    """
    matriz = []
    medias_linhas = []
    tamanho = 5

    for _ in range(tamanho):
        linha = [random.randint(1, 99) for _ in range(tamanho)]
        matriz.append(linha)

    for linha in matriz:
        media = sum(linha) / len(linha)
        medias_linhas.append(media)

    print("--- Matriz Gerada ---")
    for linha in matriz:
        print(linha)

    print("\n--- Médias das Linhas ---")
    print(medias_linhas)

calcular_media_linhas_matriz()
```

```python
# 13 -
def criar_matriz_3x3():
    """
    Solicita ao usuário para preencher uma matriz 3x3 com números inteiros.
    """
    matriz = []
    print("Digite os 9 valores para preencher a matriz 3x3:")
    for i in range(3):
        linha = []
        for j in range(3):
            while True:
                try:
                    valor = int(input(f"Digite o valor para a posição [{i+1}][{j+1}]: "))
                    linha.append(valor)
                    break
                except ValueError:
                    print("Entrada inválida. Por favor, digite um número inteiro.")
        matriz.append(linha)
    return matriz

def verificar_matriz_identidade(matriz):
    """
    Verifica se a matriz fornecida é uma matriz identidade.
    """
    for i in range(3):
        for j in range(3):
            if i == j and matriz[i][j] != 1:
                return False
            elif i != j and matriz[i][j] != 0:
                return False
    return True

def main():
    """
    Função principal que gerencia o fluxo do programa.
    """
    matriz_usuario = criar_matriz_3x3()

    print("\n--- Matriz preenchida ---")
    for linha in matriz_usuario:
        print(linha)

    if verificar_matriz_identidade(matriz_usuario):
        print("\nA matriz digitada é uma matriz identidade.")
    else:
        print("\nA matriz digitada não é uma matriz identidade.")

if __name__ == "__main__":
    main()
```

```python
# 14 -
def ler_matriz():
    """
    Lê uma matriz 3x3 com valores inteiros fornecidos pelo usuário.
    """
    matriz = []
    print("Digite os valores para a matriz 3x3:")
    for i in range(3):
        linha = []
        for j in range(3):
            while True:
                try:
                    valor = int(input(f"Digite o valor para a posição [{i+1}][{j+1}]: "))
                    linha.append(valor)
                    break
                except ValueError:
                    print("Entrada inválida. Por favor, digite um número inteiro.")
        matriz.append(linha)
    return matriz

def verificar_simetria(matriz):
    """
    Verifica se a matriz é simétrica comparando cada elemento com seu transposto.
    """
    for i in range(3):
        for j in range(i + 1, 3):
            if matriz[i][j] != matriz[j][i]:
                return False
    return True

def exibir_matriz(matriz):
    """
    Exibe a matriz de forma formatada.
    """
    print("\n--- Matriz preenchida ---")
    for linha in matriz:
        print(linha)

def main():
    """
    Função principal que gerencia a leitura e verificação da matriz.
    """
    matriz_do_usuario = ler_matriz()
    exibir_matriz(matriz_do_usuario)

    if verificar_simetria(matriz_do_usuario):
        print("\nA matriz é simétrica.")
    else:
        print("\nA matriz não é simétrica.")

if __name__ == "__main__":
    main()
```

```python
#17
def calcular_determinante_3x3(matriz):


    a, b, c = matriz[0]
    d, e, f = matriz[1]
    g, h, i = matriz[2]


    termo1 = a * e * i
    termo2 = b * f * g
    termo3 = c * d * h

    soma_diagonais_principais = termo1 + termo2 + termo3


    termo4 = c * e * g
    termo5 = a * f * h
    termo6 = b * d * i

    soma_diagonais_secundarias = termo4 + termo5 + termo6


    determinante = soma_diagonais_principais - soma_diagonais_secundarias

    return determinante


matriz_exemplo = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

det = calcular_determinante_3x3(matriz_exemplo)
print(f"O determinante da matriz é: {det}")
```

```python
#22
def obter_e_filtrar_palavras():
    """
    Solicita 10 palavras ao usuário, armazena em um conjunto
    as palavras distintas com mais de 5 letras.
    """

    palavras_digitadas = []


    print("--- Digite 10 palavras ---")
    for i in range(10):
        while True:

            palavra = input(f"Insira a palavra {i + 1} de 10: ").strip()


            if palavra:

                palavras_digitadas.append(palavra.lower())
                break
            else:
                print("Entrada inválida. Por favor, digite uma palavra.")


    palavras_filtradas = set()


    for palavra in palavras_digitadas:

        if len(palavra) > 5:

            palavras_filtradas.add(palavra)

    return palavras_filtradas


conjunto_final = obter_e_filtrar_palavras()


print("\n" + "=" * 40)
print("             RESULTADO DA FILTRAGEM")
print("=" * 40)
print(f"Total de palavras distintas com mais de 5 letras: {len(conjunto_final)}")
print("Conjunto de Palavras Filtradas:")
print(conjunto_final)
```
